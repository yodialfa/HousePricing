```python
import re
def condition(x):
    if 'M' in x:
        return "M"
    elif 'jt' in x:
        return "jt"
    else:
        return 'missing'

new_df['missing value'] = new_df['price'].apply(condition)
new_df = new_df[~(new_df['missing value']=='missing')]



def decimal(y):
#     result = re.findall('\[0-9]\.\[0-9]|', y)
#     result = re.findall('\d+\.\d+.\d+|\d+.', y)
#     result = re.findall('^\d*[.,]?\d*$', y)
    result = re.findall('\d+.\d+|\d+.', y)
    return result[0]
    

new_df['price_c'] = new_df['price'].apply(decimal)


new_df['price_c'] = new_df['price_c'].str.replace(',', '').astype(float) 
# new_df = new_df[new_df['price']!=0]

def price_result(x,y):
    if x=='jt':
        price = y * 1000000
    elif x=='M':
        price = y * 1000000000
    else:
        price = 0
    return price

new_df['normal_price'] = new_df.apply(lambda col: price_result(col['missing value'], col['price_c']),axis=1)


def number(k):
    result = re.findall('\d+.\d+.\d+', k)
    return result[0]
                        
new_df['listing-flr'] = new_df.apply(lambda col: number(col['listing-floorarea 2']),axis=1)
new_df['listing-flr'] = new_df['listing-flr'].str.replace('.','', regex=True).astype(int)
new_df
```

```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs4
```

```python
allkec = []
def scrapedata(link):
    continents_page = requests.get(link).text
    continents_countries_soup = bs4(continents_page,"html.parser")
    continents = continents_countries_soup.find_all('td', {"class":"navbox-list navbox-odd hlist"}, 'li' > 'a')
    target_continents = [continent.text for continent in continents]
    kec = target_continents[0]
    kec = kec.split(sep='\n')
    kec = [x for x in kec if x != '']
    for x in kec:
        allkec.append(x)
    return allkec

bandung = "https://id.wikipedia.org/wiki/Daftar_kecamatan_dan_kelurahan_di_Kota_Bandung"
bandungkab = "https://id.wikipedia.org/wiki/Daftar_kecamatan_dan_kelurahan_di_Kabupaten_Bandung"
cimahi = "https://id.wikipedia.org/wiki/Daftar_kecamatan_dan_kelurahan_di_Kota_Cimahi"
bdgbarat = "https://id.wikipedia.org/wiki/Daftar_kecamatan_dan_kelurahan_di_Kabupaten_Bandung_Barat"
link = [bandung,bandungkab,cimahi,bdgbarat]

for x in link:
    scrapedata(x)
    
print(allkec)
```

```python
def check_outlier(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    
    return ['outlier' if (x<q1-1.5*iqr or x>q3+1.5*iqr) 
                                 else 'not-outlier' for x in data[column].values]


final_df['outlier'] = check_outlier(data=final_df, column='normal_price')
```
